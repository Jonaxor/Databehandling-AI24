```python
import pandas as pd
import plotly.express as px


gapminder_df = px.data.gapminder()
```

```python
gapminder_df.info()
```

```python
gapminder_df["country"].unique()
```

```python
gapminder_df["country"].value_counts()
```

```python
sweden_filter = gapminder_df["country"] == "Sweden"
sweden_df = gapminder_df[sweden_filter].reset_index(drop=True)
nordic_filter = gapminder_df["country"].isin(["Sweden", "Norway", "Denmark", "Finland", "Iceland"])
nordic_df = gapminder_df[nordic_filter].reset_index(drop=True)
sweden_df
```

```python
fig = px.scatter(gapminder_df, x="gdpPercap", y="lifeExp", title="Life expectancy vs GDP per capita")
fig.show()
```

```python
fig = px.scatter(sweden_df, x="gdpPercap", y="lifeExp", title="Life expectancy vs GDP per capita")
fig.show()
```

```python
fig = px.scatter(gapminder_df, x="gdpPercap", y="lifeExp", title="Life expectancy vs GDP per capita as a bubble chart",
                 log_x=True, hover_name="country", color="continent", size="pop")
fig.show()
```

```python
fig = px.line(nordic_df, x="year", y="lifeExp", title="Nordic countries life exp over time", color="country", hover_name="pop")
fig.show()
```

```python
year1952_filter = gapminder_df["year"] == 1952

fig = px.bar(gapminder_df[year1952_filter], x="country", y="lifeExp", color="country", hover_name="pop",
            title="Life exp in world 1952", labels={"lifeExp": "Life expectancy", "country": "Countries"})
fig.show()
```

```python
fig = px.bar(nordic_df, x="year", y='lifeExp', color="country", barmode="group", hover_name="pop", title="Nordics life expover time")
fig.show()
```

```python
fig = px.scatter(gapminder_df, x="gdpPercap", y="lifeExp", hover_name="country", color="continent", size="pop", size_max=60, log_x=True)
fig.show()
```

```python
fig = px.scatter(gapminder_df, x="gdpPercap", y="lifeExp", hover_name="country", color="continent", size="pop", size_max=50, log_x=True, facet_col="continent")
fig.show()
```

```python
fig = px.scatter(nordic_df, x="gdpPercap", y="lifeExp", hover_name="country", color="year", size="pop", size_max=60, log_x=True, facet_col="country")
fig.show()
```

```python
fig = px.scatter(gapminder_df, x="year", y="lifeExp", color="continent",
                title="gdp per capita", log_x=True)
fig.show()
```

```python
fig = px.scatter(gapminder_df, x="gdpPercap", y="lifeExp", color="continent", size="pop", log_x=True, size_max=50, opacity=0.6)
fig.show()
```

```python
nordic_gdp_per_cap_max = nordic_df["gdpPercap"].max()
nordic_life_exp_max = nordic_df["lifeExp"].max()

fig = px.scatter(nordic_df, x="gdpPercap", y="lifeExp", size="pop", color="country",
                animation_frame="year", range_x=[0, nordic_gdp_per_cap_max], range_y=[60, nordic_life_exp_max + 5])
fig.show()
```

```python
fig = px.line(nordic_df, x="year", y="gdpPercap", color="country")

fig.update_layout(hovermode="x")
fig.update_xaxes(showspikes=True, spikecolor="green", spikesnap="cursor", spikemode="across", spikedash="solid")
fig.show()
```

```python
gapminder_2007_filter = gapminder_df["year"] == 2007

nordic_2007_filter = gapminder_df[gapminder_2007_filter & nordic_filter]
```

```python
px.pie(nordic_2007_filter, values="pop", names="country", title="Population of Nordic countries in 2007", hover_name="lifeExp", hover_data=["gdpPercap"])
```
