```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example tips dataset
tips_df = sns.load_dataset("tips")
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips_df)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips_df, hue="size")
```

```python

flights_df = sns.load_dataset("flights")
```

```python
sns.lineplot(x="year", y="passengers", data=flights_df)
```

```python
sns.set_style("whitegrid")
sns.set_pallette("")
sns.lineplot(x="year", y="passengers", data=flights_df, hue="month")
plt.show()
```
