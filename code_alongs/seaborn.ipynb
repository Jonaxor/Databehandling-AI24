## Introduction to Seaborn

**Seaborn** is a powerful Python library for data visualization built on top of Matplotlib.

While Matplotlib is great for creating basic plots, Seaborn makes it easier to create beautiful, informative visualizations with less effort.

Some commononly used Seaborn plots are:

* Scatter plots
* Line plots
* Bar plots
* Violin plots
* Pair plots
* Regression plots
* Historgram plots

_____

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

## Built-in Seaborn datasets

```python
tips_df = sns.load_dataset('tips')

tips_df
```

_____

## Scatter plot

```python
sns.scatterplot(x='total_bill', y='tip', data=tips_df)
```

We can add more dimensions of data!

**Add categorical differentiation**

```python
sns.scatterplot(x='total_bill', y='tip', data=tips_df, hue='sex')
```

____

## Line plots

```python
flights_df = sns.load_dataset('flights')

flights_df.head(15)
```

```python
sns.lineplot(x='year', y='passengers', data=flights_df)
plt.show()
```

**Grouping by a category**

```python
sns.lineplot(x='year', y='passengers', data=flights_df, hue='month')
plt.show()
```

_____

## Bar plots

```python
tips_df
```

```python
sns.barplot(x='day', y='total_bill', data=tips_df)
```

```python
sns.barplot(x='day', y='total_bill', hue='sex' , data=tips_df)
```

_____

## Violin plots

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips_df)
plt.show()
```

______

## Histogram



```python
sns.histplot(data=tips_df, x='total_bill', bins=20,)
```

```python
sns.histplot(data=tips_df, x='total_bill', bins=20, kde=True, hue='smoker')
```

____

## Pair plots

```python
sns.pairplot(tips_df)
```

____

## KDE Plot

```python
sns.kdeplot(data=tips_df, x='total_bill', hue='sex')
```

_____

## Settings themes



```python
sns.set_theme(style='whitegrid')

sns.scatterplot(x='total_bill', y='tip', data=tips_df, hue='sex')
```

_____

## Color Palettes

```python
sns.set_palette('bright')

sns.scatterplot(x='total_bill', y='tip', data=tips_df, hue='sex')
```

____

## Combining with Matplotlib

```python
sns.set_palette('pastel')

sns.scatterplot(x='total_bill', y='tip', data=tips_df, hue='sex')

plt.title('My title is here')
plt.xlabel('This was clear af')
plt.ylabel('woooo')
plt.show()
```
